```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Drop the original 'hsc_s' column
df.drop('hsc_s', axis=1, inplace=True)
```

```python
# Create three new columns for the three unique categories in 'degree_t'
df['degree_t_Sci&Tech'] = df['degree_t'].apply(lambda x: 1 if x == 'Sci&Tech' else 0)
df['degree_t_Comm&Mgmt'] = df['degree_t'].apply(lambda x: 1 if x == 'Comm&Mgmt' else 0)
df['degree_t_Others'] = df['degree_t'].apply(lambda x: 1 if x == 'Others' else 0)
```

```python
# Drop the original 'degree_t' column
df.drop('degree_t', axis=1, inplace=True)
```

```python
# Function to detect and remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower) & (df[column] <= upper)]
    return df
```

```python
# apply the function to each numerical column in the dataset
num_cols = ['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p', 'salary']
for col in num_cols:
    df = remove_outliers(df, col)
```
